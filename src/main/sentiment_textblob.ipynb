```python
from textblob import TextBlob
from textblob.sentiments import NaiveBayesAnalyzer
import nltk
import pandas as pd

nltk.download('movie_reviews')
nltk.download('punkt')
```

```python
df=pd.read_csv('../../Data/News/news_test.csv')
data = df["Headline"]
sentiment_textblob = []

for sentence in df['Headline']:
    

    blob = TextBlob(sentence)
    
    # decide sentiment as positive, negative and neutral
    if blob.sentiment.polarity >= 0.05 :
        sentiment_textblob.append("Positive")
        #print("Positive")
 
    elif blob.sentiment.polarity <= - 0.05 :
         sentiment_textblob.append("Negative")
        #print("Negative")
 
    else :
        sentiment_textblob.append("Neutral")
        #print("Neutral")
```

Using Movie Sentiment

```python
df=pd.read_csv('../../Data/News/news_test.csv')
data = df["Headline"]
sentiment_textblob_movie = []

for sentence in df['Headline']:
    

    blob = TextBlob(sentence, analyzer=NaiveBayesAnalyzer())
    print(blob.sentiment.classification)
    
    # decide sentiment as positive, negative and neutral
    if blob.sentiment.classification == 'pos':
        sentiment_textblob_movie.append("Positive")
        #print("Positive")
 
    elif blob.sentiment.classification == 'neg' :
         sentiment_textblob_movie.append("Negative")
        #print("Negative")
 
    else :
        sentiment_textblob_movie.append("Neutral")
        #print("Neutral")
```

```python
#sentiment_textblob_movie
#sentiment_textblob
df1 = pd.DataFrame(columns = ['sentiment_textblob_movie'])
df1["sentiment_textblob_movie"] = sentiment_textblob_movie
#df1["sentiment_textblob"] = sentiment_textblob 'sentiment_textblob'
df1.to_csv('../../Data/sentimental_textblob_2021.csv')
```

Check Accuracy 


```python
df=pd.read_csv('../../Data/Training/news_val_2021Jan_noN.csv')
data = df["Headline"]
sentiment_textblob = []

for sentence in df['Headline']:
    

    blob = TextBlob(sentence)
    
    # decide sentiment as positive, negative and neutral
    if blob.sentiment.polarity >= 0.00 :
        sentiment_textblob.append("POSITIVE")
        #print("Positive")
 
    else: # blob.sentiment.polarity <= - 0.05 :
         sentiment_textblob.append("NEGATIVE")
        #print("Negative")
 
    # else :
    #     sentiment_textblob.append("Neutral")
    #     #print("Neutral")

df["Sentiment_textblob"] = sentiment_textblob
df.to_csv("../../Data/Training/news_val_2021Jan_noN.csv")

i = 0
for index, row in df.iterrows():
    if row['Sentiment'] != row['Sentiment_textblob']:
        i = i + 1

accuracy = (209 - i)/209
print(accuracy)
```
